```python
import librosa
import numpy as np

# Load OGG audio file
audio, sr = librosa.load('audio.ogg', sr=None)  # sr=None keeps the original sample rate
```

# –í–Ω–∞—á–∞–ª–µ –ø–æ–ø—Ä–æ–±—É–µ–º –ø—Ä–∏–º–µ–Ω–∏—Ç—å –≤–µ–π–≤–ª–µ—Ç—ã –¥–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã —É–±—Ä–∞—Ç—å —à—É–º

```python
import pywt
import numpy as np

def wavelet_denoise(signal, wavelet='db8', level=4, threshold_scale=0.5):
    coeffs = pywt.wavedec(signal, wavelet, level=level)
    sigma = np.median(np.abs(coeffs[-1])) / 0.6745
    uthresh = sigma * np.sqrt(2 * np.log(len(signal))) * threshold_scale
    coeffs_thresh = [pywt.threshold(c, value=uthresh, mode='soft') for c in coeffs]
    return pywt.waverec(coeffs_thresh, wavelet)

denoised_audio = wavelet_denoise(audio)
```

```python
import io
import soundfile as sf

buffer = io.BytesIO()
denoised_audio_np = denoised_audio.astype(np.float32)
sf.write(buffer, denoised_audio_np, samplerate=sr, format='WAV')
buffer.seek(0)
Audio(buffer.read())
```

–î–ª—è —ç—Ç–æ–≥–æ –º–æ–∂–Ω–æ –ø–æ–ø—Ä–æ–±–æ–≤–∞—Ç—å –º–æ–¥–µ–ª—å encodec –æ—Ç –∑–∞–ø—Ä–µ—â—ë–Ω–Ω–æ–π –≤ –†–æ—Å—Å–∏–∏ Meta Inc. –ò–ª–∏ –ª—é–±—É—é –¥—Ä—É–≥—É—é –Ω–∞ –í–∞—à –≤—ã–±–æ—Ä. –í –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ —Ç–æ–∫–µ–Ω–∏–∑–∞—Ü–∏—è –ø–æ–º–æ–∂–µ—Ç –Ω–∞–º —Å–∂–∞—Ç—å –∞—É–¥–∏–æ.

```python
import torchaudio
import torch
from encodec import EncodecModel
from encodec.utils import convert_audio

# Load pretrained EnCodec
model = EncodecModel.encodec_model_24khz()
model.set_target_bandwidth(6.0)  # Set compression rate

# Convert your numpy audio to torch tensor
waveform = torch.from_numpy(denoised_audio).float().unsqueeze(0).unsqueeze(0)
waveform = convert_audio(waveform, sr, model.sample_rate, model.channels)

# Encode
with torch.no_grad():
    encoded_frames = model.encode(waveform)
    codes = [frame[0] for frame in encoded_frames]  # RVQ levels √ó tokens
```

# --- –ì–µ–Ω–µ—Ä–∞—Ü–∏—è –∞—É–¥–∏–æ ---
–°–≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º –ø—Ä–∏–º–µ—Ä –∞—É–¥–∏–æ –Ω–∞ —Ä—É—Å—Å–∫–æ–º —è–∑—ã–∫–µ

```python
text = "–ü—Ä–∏–≤–µ—Ç, —ç—Ç–æ –ø—Ä–∏–º–µ—Ä —Ä—É—Å—Å–∫–æ–≥–æ –∞—É–¥–∏–æ –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏."  # "Hello, this is an example of Russian audio for processing."
tts = gTTS(text=text, lang='ru')
tts.save("russian_audio.mp3")

# Load the audio
audio, sr = librosa.load("russian_audio.mp3", sr=None)
print("Original Russian Audio:")
display(Audio(audio, rate=sr))

# Apply wavelet transform (Daubechies 4 wavelet)
wavelet = 'db4'
coeffs = pywt.wavedec(audio, wavelet, level=5)

# Visualize coefficients
plt.figure(figsize=(12, 8))
for i, c in enumerate(coeffs):
    plt.subplot(len(coeffs), 1, i + 1)
    plt.plot(c)
    plt.title(f'Level {i} Coefficients' if i == 0 else f'Detail Level {i}')
plt.tight_layout()
plt.show()
```

# --- –ü—Ä–∏–º–µ—Ä: –û–±–Ω–∞—Ä—É–∂–µ–Ω–∏–µ –ø–µ—Ä–µ—Ö–æ–¥–æ–≤ ---

```python
t = np.linspace(0, 1, 1000)
signal = np.sin(2 * np.pi * 7 * t) + 0.1 * np.random.randn(1000)
signal[200:220] += 5  # Add a transient spike
```

**Fourier Transform**

```python
fft_signal = np.fft.fft(signal)
freq = np.fft.fftfreq(len(t), t[1] - t[0])
plt.figure()
plt.plot(freq[:len(freq)//2], np.abs(fft_signal[:len(freq)//2]))
plt.title('Fourier Transform (Frequency Domain)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.show()
```

## üîä Wavelet Denoising

```python
def wavelet_denoise(signal, wavelet='db1', level=3):
    coeffs = pywt.wavedec(signal, wavelet, level=level)
    sigma = np.median(np.abs(coeffs[-1])) / 0.6745
    uthresh = sigma * np.sqrt(2 * np.log(len(signal)))
    coeffs = [pywt.threshold(c, value=uthresh, mode='soft') for c in coeffs]
    return pywt.waverec(coeffs, wavelet)
```

## üß± RESD Dataset

```python
class RESDDataset(Dataset):
    def __init__(self, root_dir, sr=16000, max_len=4.0, transform=None,
                 feature_type='scalogram', apply_denoise=False):
        self.root_dir = root_dir
        self.sr = sr
        self.max_len = int(sr * max_len)
        self.transform = transform
        self.feature_type = feature_type
        self.apply_denoise = apply_denoise
        self.file_paths = []
        self.labels = []
        self._collect_files()

    def _extract_label_from_path(path):
      """Try to infer emotion label from the path string."""
      path_lower = path.lower()
      for keyword in LABEL_MAP.keys():
          if keyword in path_lower:
              return LABEL_MAP[keyword]
      return None  # Unknown


    def _collect_files(self):
        for subdir, _, files in os.walk(self.root_dir):
            for file in files:
                if file.lower().endswith('.wav'):
                    full_path = os.path.join(subdir, file)
                    label = RESDDataset._extract_label_from_path(full_path)
                    if label is not None:
                        self.file_paths.append(full_path)
                        self.labels.append(label)


    def __len__(self):
        return len(self.file_paths)

    def __getitem__(self, idx):
        path = self.file_paths[idx]
        label = self.labels[idx]
        audio, _ = librosa.load(path, sr=self.sr)

        if self.apply_denoise:
            audio = wavelet_denoise(audio)

        if len(audio) < self.max_len:
            audio = np.pad(audio, (0, self.max_len - len(audio)))
        else:
            audio = audio[:self.max_len]

        if self.feature_type == 'scalogram':
            features = self._get_scalogram(audio)
        else:
            raise ValueError("Use 'scalogram' as feature_type")

        return features, label

    def _get_scalogram(self, signal):
        scales = np.arange(1, 64)
        coeffs, _ = pywt.cwt(signal, scales, 'morl')
        power = np.abs(coeffs)
        power = power[:64, :256]  # Crop/pad to fixed size
        return torch.tensor(power).float()
```

## üß† CNN Model

–ö–ª–∞—Å—Å–∏—á–µ—Å–∫–∞—è CNN, –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç –æ–±—É—á–∞—Ç—å—Å—è "–≤ –ª–æ–±", –¥–∞–∂–µ –µ—Å–ª–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å—Ç–∞–Ω–æ–≤—è—Ç—Å—è —Ö—É–∂–µ.

```python
class EmotionCNN(nn.Module):
    def __init__(self, num_classes=7):
        super().__init__()
        self.conv1 = nn.Conv2d(1, 16, 3, padding=1)
        self.conv2 = nn.Conv2d(16, 32, 3, padding=1)
        self.conv3 = nn.Conv2d(32, 64, 3, padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        self.dropout = nn.Dropout(0.3)
        # Adjusted input size for fc1 based on calculated flattened dimensions (64 * 7 * 32)
        self.fc1 = nn.Linear(64 * 7 * 32, 128)
        self.fc2 = nn.Linear(128, num_classes)

    def forward(self, x):
        x = x.unsqueeze(1)  # (B, 1, H, W) where H=63, W=256
        x = self.pool(F.relu(self.conv1(x))) # (B, 16, 31, 128)
        x = self.pool(F.relu(self.conv2(x))) # (B, 32, 15, 64)
        x = self.pool(F.relu(self.conv3(x))) # (B, 64, 7, 32)
        x = x.view(x.size(0), -1) # Flatten to (B, 64 * 7 * 32)
        x = self.dropout(F.relu(self.fc1(x)))
        return self.fc2(x)
```

–ß—Ç–æ–±—ã —Å—ç–∫–æ–Ω–æ–º–∏—Ç—å –≤—Ä–µ–º—è –Ω–∞ –æ–±—É—á–µ–Ω–∏–µ –Ω–µ–π—Ä–æ–Ω–Ω–æ–π —Å–µ—Ç–∏ (–æ–∫–æ–ª–æ 2—Ö —á–∞—Å–æ–≤), –º–æ–∂–Ω–æ –ø—Ä–æ—Å—Ç–æ –∑–∞–≥—Ä—É–∑–∏—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã, –∫–æ—Ç–æ—Ä—ã–µ –ø–æ–ª—É—á–∏–ª–∏—Å—å —É –º–µ–Ω—è. –î–∞, —ç—Ç–æ –Ω–µ —Å–∞–º—ã–µ –ª—É—á—à–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã, –∫–æ—Ç–æ—Ä—ã–µ –Ω–∏ –Ω–∞ —á—Ç–æ –æ—Å–æ–±–µ–Ω–Ω–æ–µ –Ω–µ –ø—Ä–µ—Ç–µ–Ω–¥—É—é—Ç, –Ω–æ –í—ã –≤—Å–µ–≥–¥–∞ –º–æ–∂–µ—Ç–µ –∏—Ö —É–ª—É—á—à–∏—Ç—å, –∫–æ–≥–¥–∞ —É –í–∞—Å –±—É–¥–µ—Ç –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –≤—Ä–µ–º–µ–Ω–∏!

–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–æ –º–æ–∂–Ω–æ —Å–æ–∑–¥–∞—Ç—å `EarlyStopping` CNN, –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç –∑–∞–≤–µ—Ä—à–∞—Ç—å –æ–±—É—á–µ–Ω–∏–µ, –µ—Å–ª–∏ –≤—Å—ë –∏–¥—ë—Ç –Ω–µ –ø–æ –ø–ª–∞–Ω—É.

```python
import pytorch_lightning as pl
import torch.nn.functional as F

class EmotionClassifier(pl.LightningModule):
    def __init__(self, num_classes=7, lr=1e-3):
        super().__init__()
        self.save_hyperparameters()
        self.cnn = EmotionCNN(num_classes)

    def forward(self, x):
        return self.cnn(x)

    def training_step(self, batch, batch_idx):
        x, y = batch
        logits = self(x)
        loss = F.cross_entropy(logits, y)
        acc = (logits.argmax(dim=1) == y).float().mean()
        self.log("train_loss", loss)
        self.log("train_acc", acc)
        return loss

    def validation_step(self, batch, batch_idx):
        x, y = batch
        logits = self(x)
        loss = F.cross_entropy(logits, y)
        acc = (logits.argmax(dim=1) == y).float().mean()
        self.log("val_loss", loss, prog_bar=True)
        self.log("val_acc", acc, prog_bar=True)

    def configure_optimizers(self):
        return torch.optim.Adam(self.parameters(), lr=self.hparams.lr)
```

## üèãÔ∏è Train Model

–û–±—É—á–µ–Ω–∏–µ –¥–ª—è –∫–ª–∞—Å—Å–∏—á–µ—Å–∫–æ–π CNN **(–∑–∞–Ω–∏–º–∞–µ—Ç 2 —á–∞—Å–∞, –Ω–µ –¥–µ–ª–∞–π—Ç–µ —ç—Ç–æ –ø–æ–∂–∞–ª—É–π—Å—Ç–∞ –Ω–∞ –ø–∞—Ä–µ).**

```python
model = EmotionCNN().to(torch.device('cuda' if torch.cuda.is_available() else 'cpu'))
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)

train_losses, val_losses, val_accuracies = [], [], []
print(datetime.datetime.now())
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

for epoch in range(10):
    model.train()
    running_loss = 0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

    train_losses.append(running_loss / len(train_loader))

    # Validation
    model.eval()
    val_loss, correct, total = 0, 0, 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            val_loss += loss.item()
            _, preds = torch.max(outputs, 1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

    val_losses.append(val_loss / len(val_loader))
    val_accuracies.append(correct / total)

    print(f"{datetime.datetime.now()} Epoch {epoch+1}: Train Loss={train_losses[-1]:.4f}, Val Acc={val_accuracies[-1]:.4f}")
```

```python
np.random.seed(0)
t = np.linspace(0, 1, 1000)
speech_segments = [np.sin(2 * np.pi * 1 * t) + 0.1 * np.random.randn(1000) for _ in range(20)]
music_segments = [np.sin(2 * np.pi * 10 * t) + 0.1 * np.random.randn(1000) for _ in range(20)]
labels = [0] * 20 + [1] * 20  # 0: speech, 1: music
```

**Extract MFCC features**

```python
def extract_mfcc(segment, sr=1000):
    mfcc = librosa.feature.mfcc(y=segment, sr=sr, n_mfcc=13)
    return np.mean(mfcc, axis=1)

mfcc_features = [extract_mfcc(seg) for seg in speech_segments + music_segments]
```

**Extract wavelet features**

```python
def extract_wavelet_features(segment, wavelet='db4', level=5):
    coeffs = pywt.wavedec(segment, wavelet, level=level)
    return [np.mean(abs(c)) for c in coeffs]

wavelet_features = [extract_wavelet_features(seg) for seg in speech_segments + music_segments]
```

**Train classifiers**

```python
X_mfcc = np.array(mfcc_features)
X_wavelet = np.array(wavelet_features)
y = np.array(labels)

X_train_mfcc, X_test_mfcc, y_train, y_test = train_test_split(X_mfcc, y, test_size=0.2, random_state=0)
X_train_wavelet, X_test_wavelet, _, _ = train_test_split(X_wavelet, y, test_size=0.2, random_state=0)

clf_mfcc = LogisticRegression(max_iter=1000)
clf_mfcc.fit(X_train_mfcc, y_train)
y_pred_mfcc = clf_mfcc.predict(X_test_mfcc)
print(f"Accuracy with MFCC: {accuracy_score(y_test, y_pred_mfcc):.2f}")

clf_wavelet = LogisticRegression(max_iter=1000)
clf_wavelet.fit(X_train_wavelet, y_train)
y_pred_wavelet = clf_wavelet.predict(X_test_wavelet)
print(f"Accuracy with Wavelet Features: {accuracy_score(y_test, y_pred_wavelet):.2f}")
```
